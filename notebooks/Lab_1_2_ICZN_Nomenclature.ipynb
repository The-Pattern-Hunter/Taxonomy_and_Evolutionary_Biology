## Part 1: ICZN Theory

### International Code of Zoological Nomenclature

**Purpose:** Ensure every animal species has ONE unique scientific name

### Key Principles

**1. Binomial Nomenclature (Linnaeus, 1758)**
- Every species has TWO names
- Genus + species epithet
- Example: *Homo sapiens*

**2. Priority Rule**
- Oldest validly published name wins
- Starting point: 1758 (Linnaeus, Systema Naturae, 10th ed.)
- First description = valid name

**3. Type Specimens**
- Every species based on physical specimen(s)
- Holotype: Single designated specimen
- Paratypes: Additional specimens from original series
- Name attached to type, not concept

**4. Publication Requirements**
- Must be printed and distributed
- Must contain description or diagnosis
- Must designate type specimen
- Must be publicly available

### Name Format Rules

**Genus:**
- Capitalized
- Italicized (or underlined)
- Singular noun
- Example: *Felis*

**Species epithet:**
- Lowercase
- Italicized
- Can be adjective, noun, or latinized word
- Example: *catus*

**Complete name:**
- *Felis catus* Linnaeus, 1758
- Authority (author + year) optional but recommended

**Subspecies (trinomial):**
- *Panthera tigris tigris* (Bengal tiger)
- Three names: Genus + species + subspecies

### Common Problems

**Synonymy:**
- Same species named multiple times
- Senior synonym (older) = valid
- Junior synonym (newer) = invalid
- Example: *Bos taurus* = *Bos primigenius taurus*

**Homonymy:**
- Same name used for different species
- Senior homonym keeps name
- Junior homonym must be renamed
- Example: *Myrmica sulcinodis* (ant) vs *Myrmica sulcinodis* (wasp) - wasp renamed!

**Changes in Classification:**
- Species moved to different genus
- Original authority in parentheses
- Example: *Apis mellifera* (Linnaeus, 1758)
- Originally: *Apis* â†’ stays *Apis*
- If moved: *Bombus mellifera* (Linnaeus, 1758)

### ICZN Operative Principles

**Article 23: Priority**
- Valid name = oldest available name
- Exceptions: Conservation (rare)

**Article 51: Homonymy**
- Primary homonym: Same genus and species
- Secondary homonym: After genus transfer
- Junior homonym permanently invalid

**Article 61: Synonymy**
- Objective synonym: Same type specimen
- Subjective synonym: Different types, same species (opinion)

**Article 72-74: Name Formation**
- Correct latinization
- Gender agreement
- Mandatory changes vs corrections

## Part 3: Priority Rule Simulator

```python
def resolve_priority(names_list):
    """
    Apply priority rule to list of names
    names_list: list of dicts with 'name', 'author', 'year'
    """
    # Sort by year
    sorted_names = sorted(names_list, key=lambda x: x['year'])
    
    # Identify valid (senior) name
    valid_name = sorted_names[0]
    junior_synonyms = sorted_names[1:]
    
    # Visualization
    fig = go.Figure()
    
    # Timeline
    years = [n['year'] for n in sorted_names]
    names = [f"{n['name']}<br>{n['author']}" for n in sorted_names]
    colors = ['green' if n == valid_name else 'red' for n in sorted_names]
    
    fig.add_trace(go.Scatter(
        x=years,
        y=[1]*len(years),
        mode='markers+text',
        marker=dict(size=30, color=colors),
        text=names,
        textposition='top center',
        showlegend=False
    ))
    
    fig.update_layout(
        title='<b>Priority Rule Timeline</b>',
        xaxis_title='Year of Publication',
        yaxis=dict(showticklabels=False, range=[0, 2]),
        height=400
    )
    
    # Results
    print("\n" + "="*70)
    print("PRIORITY RULE APPLICATION")
    print("="*70)
    print(f"\nALL NAMES (chronological):")
    for i, name in enumerate(sorted_names, 1):
        status = "âœ“ VALID (senior synonym)" if name == valid_name else "âœ— INVALID (junior synonym)"
        print(f"  {i}. {name['name']} {name['author']}, {name['year']} - {status}")
    
    print(f"\nPRIORITY RULE (Article 23):")
    print(f"  â†’ Valid name = oldest available name")
    print(f"  â†’ All later names are junior synonyms")
    
    print(f"\nâœ“ VALID NAME: {valid_name['name']} {valid_name['author']}, {valid_name['year']}")
    
    if junior_synonyms:
        print(f"\nâœ— JUNIOR SYNONYMS (invalid):")
        for syn in junior_synonyms:
            print(f"  = {syn['name']} {syn['author']}, {syn['year']}")
    
    print(f"\nKEY POINT:")
    print(f"  All names refer to SAME species")
    print(f"  Priority determines which name is used")
    print(f"  Junior synonyms permanently invalid")
    print("="*70)
    
    fig.show()

# Example: Tiger names
tiger_names = [
    {'name': 'Felis tigris', 'author': 'Linnaeus', 'year': 1758},
    {'name': 'Tigris striata', 'author': 'Severtzov', 'year': 1858},
    {'name': 'Tigris regalis', 'author': 'Gray', 'year': 1867}
]

display(HTML("<h3>âš–ï¸ Priority Rule: Tiger Names</h3>"))
resolve_priority(tiger_names)
```

## Part 5: Name Formatting Checker

```python
def check_name_format(genus, species, author=None, year=None, genus_changed=False):
    """
    Check if scientific name is correctly formatted
    """
    errors = []
    warnings = []
    
    # Check genus
    if not genus[0].isupper():
        errors.append("Genus must be capitalized")
    if genus.islower():
        errors.append("Genus should have capital first letter only")
    
    # Check species epithet
    if species[0].isupper():
        errors.append("Species epithet must be lowercase")
    
    # Check author formatting
    if author and year:
        if genus_changed and '(' not in str(author):
            warnings.append("Authority should be in parentheses (genus changed)")
        if not genus_changed and '(' in str(author):
            warnings.append("Authority should NOT be in parentheses (genus unchanged)")
    
    # Format the name
    if author and year:
        if genus_changed:
            formatted = f"*{genus} {species}* ({author}, {year})"
        else:
            formatted = f"*{genus} {species}* {author}, {year}"
    else:
        formatted = f"*{genus} {species}*"
    
    # Results
    print("\n" + "="*70)
    print("NAME FORMAT CHECKER")
    print("="*70)
    print(f"\nGENUS: {genus}")
    print(f"SPECIES: {species}")
    if author:
        print(f"AUTHOR: {author}")
    if year:
        print(f"YEAR: {year}")
    if genus_changed:
        print(f"GENUS CHANGED: Yes")
    
    print(f"\nFORMATTED NAME:")
    print(f"  {formatted}")
    print(f"  (Note: * represents italics)")
    
    if errors:
        print(f"\nâŒ ERRORS:")
        for error in errors:
            print(f"  â€¢ {error}")
    else:
        print(f"\nâœ“ NO ERRORS")
    
    if warnings:
        print(f"\nâš ï¸  WARNINGS:")
        for warning in warnings:
            print(f"  â€¢ {warning}")
    
    print(f"\nFORMATTING RULES:")
    print(f"  1. Genus: Capitalized, italicized")
    print(f"  2. Species epithet: Lowercase, italicized")
    print(f"  3. Authority: Roman (not italic)")
    print(f"  4. Parentheses: Use if genus changed from original")
    print(f"  5. Year: Optional but recommended")
    print("="*70)
    
    return formatted, errors, warnings

# Interactive checker
genus_input = Text(value='Homo', description='Genus:')
species_input = Text(value='sapiens', description='Species:')
author_input = Text(value='Linnaeus', description='Author:')
year_input = IntText(value=1758, description='Year:')
changed_check = Checkbox(value=False, description='Genus changed from original?')

display(HTML("<h3>âœï¸ Name Format Checker</h3>"))
interact(check_name_format,
        genus=genus_input,
        species=species_input,
        author=author_input,
        year=year_input,
        genus_changed=changed_check);
```

- Genus + species + subspecies
- Nominotypical subspecies repeats species name

**Key distinctions:**

HOMONYMY: Same name, different taxa (rename junior)
SUBSPECIES: Related populations (all names valid)

## Summary

### Key Concepts

âœ… **ICZN Code:** Rules governing animal nomenclature

âœ… **Type Specimens:** Name attached to physical specimen  
âœ… **Homonymy:** Same name, different species (rename junior)

### ICZN Core Principles

**1. Binomial Nomenclature**
- Genus + species epithet
- *Homo sapiens*
- Universal system since 1758

**2. Priority (Article 23)**
- Valid name = oldest available
- Starting point: Linnaeus 1758
- Junior synonyms invalid

**3. Type Specimens**
- Holotype: Single designated specimen
- Paratypes: Additional original specimens
- Name attached to type, not concept

**4. Publication**
- Printed and distributed
- Contains description
- Designates type
- Publicly available

### Name Formatting Rules

**Basic Format:**
*Genus species* Author, Year

**Example:**
*Homo sapiens* Linnaeus, 1758

**Rules:**
- Genus: Capitalized, italic
- Species: Lowercase, italic  
- Authority: Roman, capitalized
- Year: After comma

**Parentheses:**
*Panthera tigris* (Linnaeus, 1758)
- Use if genus changed from original
- Shows nomenclatural history
- Authority in parentheses too

**Subspecies (Trinomial):**
*Panthera tigris tigris* (Bengal tiger)
- Three names: Genus + species + subspecies
- Nominotypical subspecies repeats species name

### Problem Types

**Synonymy:**
- Multiple names â†’ ONE species
- Same thing, different names
- Solution: Use oldest (senior synonym)
- Junior synonyms = invalid but preserved in literature

**Homonymy:**
- One name â†’ MULTIPLE species
- Different things, same name
- Solution: Rename younger (junior homonym)
- Primary: Same genus originally
- Secondary: After genus transfer

**Example - Synonymy:**
- *Felis tigris* Linnaeus, 1758 âœ“
- = *Tigris striata* Severtzov, 1858 âœ—
- = *Tigris regalis* Gray, 1867 âœ—
- All refer to tiger

**Example - Homonymy:**
- *Noctua pronuba* (moth) 1758 âœ“
- *Noctua pronuba* (butterfly) 1790 âœ— â†’ RENAME
- Different animals, same name

### Type Specimens

**Types:**
- Holotype: THE specimen
- Paratypes: Additional from type series
- Lectotype: Selected from syntypes
- Neotype: Replacement if holotype lost

**Importance:**
- Name attached to specimen, not concept
- Permanent reference
- Resolves disputes
- Stored in museums

**If holotype lost:**
- Designate neotype
- Must be from same locality
- Maintains name stability

### ICZN Articles

**Article 23 - Priority:**
Valid name = oldest available name

**Article 51 - Homonymy:**
Junior homonym permanently invalid

**Article 61 - Synonymy:**
- Objective: Same type
- Subjective: Different types, opinion

**Article 72-74 - Formation:**
- Correct latinization
- Gender agreement
- Mandatory vs recommended changes

### Real-World Impact

**Stability:**
- One name per species (globally)
- Clear communication
- Prevents confusion

**Problems:**
- Name changes confusing to public
- Economic impact (e.g., pest species)
- Historical literature uses old names

**Solutions:**
- ICZN Commission can conserve names
- Synonym lists maintained
- Databases track changes

### The Big Picture

**ICZN ensures:**
- Every species has ONE valid name
- Names are stable and universal
- Rules are clear and applied consistently
- Disputes can be resolved

**Without ICZN:**
- Multiple names for same species
- Same name for different species
- Chaos and confusion
- No scientific communication

**Names are:**
- Hypotheses (can change with new evidence)
- But governed by rules (not arbitrary)
- Balancing stability vs accuracy
