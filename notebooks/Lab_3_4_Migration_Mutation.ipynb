## Part 1: Migration & Mutation Theory

### Migration (Gene Flow)

**Definition:**
Movement of individuals (and their alleles) between populations.

**Effect:**
Homogenizes allele frequencies across populations (OPPOSES differentiation).

**Migration rate (m):**
Proportion of individuals in a population that are immigrants each generation.

**Change in allele frequency:**
## Δq = m(q_m - q)

Where:
- m = migration rate (0 to 1)
- q_m = frequency in migrants (source population)
- q = frequency in residents (recipient population)
- Δq = change in recipient population

**Key insights:**
- If q_m > q: Δq positive (frequency increases)
- If q_m < q: Δq negative (frequency decreases)
- If q_m = q: Δq = 0 (no change)
- Larger difference (q_m - q) → larger change
- Higher migration rate → larger change

**Example:**
```
Population 1: q₁ = 0.2 (20%)
Population 2: q₂ = 0.8 (80%)
Migration: m = 0.1 (10% of pop1 are immigrants from pop2)

Δq₁ = 0.1(0.8 - 0.2) = 0.1(0.6) = 0.06
New q₁ = 0.2 + 0.06 = 0.26 (26%)

One generation of 10% migration → 6% frequency change!
```

**Equilibrium:**
At equilibrium, all populations have same frequency: q_eq = average of all populations (weighted by size).

### Migration Models

**1. Island Model:**
- Many equal-sized populations
- Each exchanges migrants with all others equally
- Homogenizes frequencies rapidly

**2. Stepping-Stone Model:**
- Populations arranged in space (1D or 2D)
- Migration only between adjacent populations
- Slower homogenization
- Creates geographic clines

**3. Continent-Island Model:**
- Large source population (continent)
- Small recipient population (island)
- One-way migration
- Island frequency approaches continent frequency

### Migration vs Other Forces

**Migration vs Selection:**
Migration can maintain locally maladaptive alleles!

If migration rate m > s (selection coefficient), gene flow wins.

**Example:** Industrial melanism
- Dark moths favored in polluted areas (s ≈ 0.3)
- But migration from rural areas (light moths) limits adaptation
- Dark moths never reach 100% if m > 0

**Migration vs Drift:**
Even small migration prevents differentiation by drift!

Rule of thumb: **Just 1 migrant per generation** (Nm = 1) is enough to prevent fixation of different alleles by drift.

### Mutation

**Definition:**
Change in DNA sequence creating new alleles.

**Mutation rate (μ):**
Probability of mutation per gene per generation.

Typical values:
- Per base pair: μ ≈ 10⁻⁹ per generation
- Per gene: μ ≈ 10⁻⁶ to 10⁻⁵ per generation

**Change in allele frequency:**

**Forward mutation** (A → a at rate μ):
## Δq = μp

**Reverse mutation** (a → A at rate ν):
## Δq = -νq

**Both directions:**
## Δq = μp - νq

**Equilibrium frequency** (mutation alone):
## q̂ = μ / (μ + ν)

**Key insight:**
Mutation is VERY SLOW! Takes hundreds of thousands of generations to significantly change allele frequencies.

### Mutation-Selection Balance

**The Question:**
Why do deleterious alleles persist despite selection against them?

**The Answer:**
Mutation keeps creating them!

**For recessive lethal** (s = 1):
## q̂ = √(μ/s) = √μ

**For general recessive** (selection s):
## q̂ = √(μ/s)

**Example:**
```
Mutation rate: μ = 10⁻⁵
Selection: s = 0.01 (1% disadvantage)

Equilibrium: q̂ = √(10⁻⁵/0.01) = √0.001 = 0.032 (3.2%)

Despite selection, allele maintained at 3%!
```

**Implications:**
- Can't eliminate genetic disease by selection alone
- Mutation creates "genetic load"
- Stronger selection → lower equilibrium frequency
- Higher mutation rate → higher equilibrium frequency

### Comparing All Forces

**Summary of effects per generation:**

| Force | Equation | Typical Magnitude |
|-------|----------|------------------|
| **Selection** | Δq ≈ spq | 10⁻³ to 10⁻¹ |
| **Drift** | σ(Δq) = √(pq/2N) | 10⁻² (N=100) to 10⁻³ (N=1000) |
| **Migration** | Δq = m(q_m - q) | 10⁻³ to 10⁻¹ |
| **Mutation** | Δq = μp | 10⁻⁶ to 10⁻⁵ |

**Ranking (fastest to slowest):**
1. **Selection** (if strong)
2. **Migration** (if high)
3. **Drift** (if small N)
4. **Mutation** (always slowest)

**Key insight:**
Mutation is TOO SLOW to change frequencies directly, BUT:
- Provides raw material for evolution (new alleles)
- Maintains variation (balances selection)
- Enables long-term evolution

### Multiple Forces Acting Together

**Net change:**
## Δq_total = Δq_selection + Δq_migration + Δq_mutation + Δq_drift

**Which force dominates?**

Depends on:
- Selection coefficient (s)
- Population size (N)
- Migration rate (m)
- Mutation rate (μ)

**Common scenarios:**

**Large population, no migration:**
- Selection dominates (drift weak)
- Allele frequencies determined by fitness
- Mutation maintains variation

**Small population, no migration:**
- Drift dominates (unless s very large)
- Random fixation/loss
- Selection less effective

**High migration:**
- Gene flow dominates
- Populations remain similar
- Local adaptation prevented

**Multiple populations, low migration:**
- Selection + drift cause differentiation
- Migration partially homogenizes
- Balance creates geographic structure

### The Big Picture

**Four fundamental forces of evolution:**

**INCREASE variation:**
- Mutation (creates new alleles)
- Migration (introduces alleles from other populations)

**DECREASE variation:**
- Drift (random loss of alleles)
- Selection (removes deleterious alleles)

**Deterministic (predictable):**
- Selection (toward higher fitness)
- Migration (toward homogenization)
- Mutation (very slowly)

**Stochastic (random):**
- Drift (sampling error)

**Result:**
Evolution is the combined action of all four forces, with their relative importance depending on population parameters!

## Part 2: Migration Calculator

````python
def migration_calculator(q_resident, q_migrant, migration_rate, generations):
    """
    Calculate change in allele frequency from gene flow
    """
    
    # Simulate over generations
    q_trajectory = [q_resident]
    
    q = q_resident
    for gen in range(generations):
        delta_q = migration_rate * (q_migrant - q)
        q = q + delta_q
        q_trajectory.append(q)
    
    # Calculate equilibrium (if migration continues)
    # At equilibrium: q_eq = q_migrant (resident matches source)
    q_eq = q_migrant
    
    # Time to 95% equilibrium
    # q(t) = q_m - (q_m - q_0)(1-m)^t
    # Solve for 95%: 0.95 = 1 - (1-m)^t
    # (1-m)^t = 0.05
    # t = ln(0.05) / ln(1-m)
    if migration_rate > 0 and migration_rate < 1:
        t_95 = np.log(0.05) / np.log(1 - migration_rate)
    else:
        t_95 = float('inf')
    
    # Visualization
    fig = make_subplots(
        rows=1, cols=2,
        subplot_titles=('Allele Frequency Over Time', 'Migration Effect per Generation'),
        specs=[[{'type': 'scatter'}, {'type': 'bar'}]]
    )
    
    gens = list(range(len(q_trajectory)))
    
    # 1. Trajectory
    fig.add_trace(go.Scatter(
        x=gens, y=q_trajectory,
        mode='lines+markers',
        line=dict(color='#3498DB', width=3),
        name='Resident population'
    ), row=1, col=1)
    
    # Add migrant frequency line
    fig.add_trace(go.Scatter(
        x=[0, generations], y=[q_migrant, q_migrant],
        mode='lines',
        line=dict(color='#E74C3C', width=2, dash='dash'),
        name='Migrant frequency'
    ), row=1, col=1)
    
    # 2. Change per generation (first 5 gens)
    delta_qs = [q_trajectory[i+1] - q_trajectory[i] for i in range(min(5, generations))]
    fig.add_trace(go.Bar(
        x=[f"Gen {i+1}" for i in range(len(delta_qs))],
        y=delta_qs,
        marker_color='#2ECC71',
        showlegend=False
    ), row=1, col=2)
    
    fig.update_xaxes(title_text="Generation", row=1, col=1)
    fig.update_yaxes(title_text="Frequency (q)", range=[0, 1], row=1, col=1)
    fig.update_yaxes(title_text="Δq", row=1, col=2)
    
    fig.update_layout(height=400, title_text='<b>Gene Flow Simulation</b>')
    
    # Print results
    print("\n" + "="*70)
    print("MIGRATION (GENE FLOW) ANALYSIS")
    print("="*70)
    print(f"\nINITIAL CONDITIONS:")
    print(f"  Resident population q: {q_resident:.3f}")
    print(f"  Migrant population q_m: {q_migrant:.3f}")
    print(f"  Difference (q_m - q): {q_migrant - q_resident:+.3f}")
    print(f"  Migration rate (m): {migration_rate:.3f} ({migration_rate*100:.1f}%)")
    print(f"\nFIRST GENERATION:")
    delta_q_first = migration_rate * (q_migrant - q_resident)
    print(f"  Δq = m(q_m - q) = {migration_rate:.3f} × {q_migrant - q_resident:.3f} = {delta_q_first:+.4f}")
    print(f"  New q = {q_resident:.3f} + {delta_q_first:.4f} = {q_resident + delta_q_first:.3f}")
    print(f"\nAFTER {generations} GENERATIONS:")
    print(f"  Final q: {q_trajectory[-1]:.3f}")
    print(f"  Total change: {q_trajectory[-1] - q_resident:+.3f}")
    print(f"  % toward equilibrium: {abs(q_trajectory[-1] - q_resident)/abs(q_migrant - q_resident)*100:.1f}%")
    print(f"\nEQUILIBRIUM:")
    print(f"  q_equilibrium: {q_eq:.3f} (matches migrant frequency)")
    print(f"  Time to 95% equilibrium: {t_95:.1f} generations")
    print(f"\nKEY INSIGHTS:")
    if migration_rate > 0.1:
        print(f"  → HIGH migration rate (>10%)")
        print(f"  → Rapid homogenization")
        print(f"  → Gene flow dominates other forces")
    elif migration_rate > 0.01:
        print(f"  → MODERATE migration rate (1-10%)")
        print(f"  → Gradual homogenization")
    else:
        print(f"  → LOW migration rate (<1%)")
        print(f"  → Slow homogenization")
        print(f"  → Local adaptation possible")
    
    # Compare to other forces
    print(f"\nCOMPARISON TO OTHER FORCES (first generation):")
    print(f"  Migration: Δq = {abs(delta_q_first):.4f}")
    print(f"  Selection (s=0.1): Δq ≈ {0.1 * q_resident * (1-q_resident):.4f}")
    print(f"  Drift (N=100): σ(Δq) ≈ {np.sqrt(q_resident*(1-q_resident)/200):.4f}")
    print(f"  Mutation (μ=10⁻⁵): Δq ≈ {1e-5 * (1-q_resident):.6f}")
    print("="*70)
    
    fig.show()

# Interactive calculator
q_res_slider = FloatSlider(value=0.2, min=0.1, max=0.9, step=0.1,
                          description='Resident q:')
q_mig_slider = FloatSlider(value=0.8, min=0.1, max=0.9, step=0.1,
                          description='Migrant q:')
m_slider = FloatSlider(value=0.1, min=0.01, max=0.5, step=0.01,
                      description='Migration (m):')
gen_slider = IntSlider(value=50, min=10, max=200, step=10,
                      description='Generations:')

display(HTML("<h3>🌊 Migration Calculator</h3>"))
interact(migration_calculator, q_resident=q_res_slider, q_migrant=q_mig_slider,
        migration_rate=m_slider, generations=gen_slider);
````

## Part 3: Mutation-Selection Balance Calculator

````python
def mutation_selection_balance(mutation_rate, selection_coeff, dominance='recessive'):
    """
    Calculate equilibrium frequency under mutation-selection balance
    """
    
    # Calculate equilibrium
    if dominance == 'recessive':
        # q_hat = sqrt(mu/s)
        q_eq = np.sqrt(mutation_rate / selection_coeff)
        formula = "√(μ/s)"
    elif dominance == 'dominant':
        # q_hat ≈ mu/s (for small mu)
        q_eq = mutation_rate / selection_coeff
        formula = "μ/s"
    else:  # additive
        # q_hat ≈ mu/s
        q_eq = mutation_rate / selection_coeff
        formula = "μ/s"
    
    # Calculate genotype frequencies at equilibrium
    p_eq = 1 - q_eq
    freq_AA = p_eq**2
    freq_Aa = 2 * p_eq * q_eq
    freq_aa = q_eq**2
    
    # Genetic load
    if dominance == 'recessive':
        L = selection_coeff * q_eq**2
    else:
        L = selection_coeff * (2 * p_eq * q_eq + q_eq**2)
    
    # Time to equilibrium (rough estimate)
    # t ≈ 1/mu for mutation-dominated
    t_eq_mutation = int(1 / mutation_rate) if mutation_rate > 0 else float('inf')
    
    # Visualization
    fig = make_subplots(
        rows=1, cols=2,
        subplot_titles=('Equilibrium Genotype Frequencies',
                       'Balance of Forces'),
        specs=[[{'type': 'bar'}, {'type': 'scatter'}]]
    )
    
    # 1. Genotype frequencies
    genotypes = ['AA', 'Aa', 'aa']
    frequencies = [freq_AA, freq_Aa, freq_aa]
    colors = ['#3498DB', '#2ECC71', '#E74C3C']
    
    fig.add_trace(go.Bar(
        x=genotypes, y=frequencies,
        marker_color=colors,
        text=[f"{f:.4f}" for f in frequencies],
        textposition='outside'
    ), row=1, col=1)
    
    # 2. Forces over allele frequency range
    q_range = np.linspace(0.001, 0.2, 100)
    mutation_force = [mutation_rate * (1-q) for q in q_range]
    if dominance == 'recessive':
        selection_force = [selection_coeff * q**2 for q in q_range]
    else:
        selection_force = [selection_coeff * q * (2*(1-q) + q) for q in q_range]
    
    fig.add_trace(go.Scatter(
        x=q_range, y=mutation_force,
        mode='lines', name='Mutation (↑)',
        line=dict(color='#F39C12', width=2)
    ), row=1, col=2)
    
    fig.add_trace(go.Scatter(
        x=q_range, y=selection_force,
        mode='lines', name='Selection (↓)',
        line=dict(color='#E74C3C', width=2)
    ), row=1, col=2)
    
    # Mark equilibrium
    if q_eq < 0.2:
        fig.add_vline(x=q_eq, line_dash="dash", line_color="green",
                     annotation_text="Equilibrium", row=1, col=2)
    
    fig.update_yaxes(title_text="Frequency", row=1, col=1)
    fig.update_xaxes(title_text="q (allele frequency)", row=1, col=2)
    fig.update_yaxes(title_text="Force strength", row=1, col=2)
    
    fig.update_layout(height=400,
                     title_text='<b>Mutation-Selection Balance</b>')
    
    # Print results
    print("\n" + "="*70)
    print("MUTATION-SELECTION BALANCE")
    print("="*70)
    print(f"\nPARAMETERS:")
    print(f"  Mutation rate (μ): {mutation_rate:.2e}")
    print(f"  Selection coefficient (s): {selection_coeff:.3f}")
    print(f"  Dominance: {dominance}")
    print(f"\nEQUILIBRIUM:")
    print(f"  Formula: q̂ = {formula}")
    print(f"  q̂ = {q_eq:.6f} ({q_eq*100:.4f}%)")
    print(f"  p̂ = {p_eq:.6f}")
    print(f"\nGENOTYPE FREQUENCIES AT EQUILIBRIUM:")
    print(f"  AA: {freq_AA:.6f} ({freq_AA*100:.4f}%)")
    print(f"  Aa: {freq_Aa:.6f} ({freq_Aa*100:.4f}%)")
    print(f"  aa: {freq_aa:.6f} ({freq_aa*100:.6f}%)")
    print(f"\nDISEASE INCIDENCE (if aa is affected):")
    print(f"  Frequency: {freq_aa:.2e} = 1 in {int(1/freq_aa) if freq_aa > 0 else 'infinite'}")
    if dominance == 'recessive':
        carrier_freq = freq_Aa
        print(f"  Carrier frequency: {carrier_freq:.2e} = 1 in {int(1/carrier_freq) if carrier_freq > 0 else 'infinite'}")
        if freq_aa > 0:
            ratio = carrier_freq / freq_aa
            print(f"  Carrier/affected ratio: {ratio:.1f}:1")
    print(f"\nGENETIC LOAD:")
    print(f"  L = {L:.6f}")
    print(f"  Mean fitness: w̄ = {1-L:.6f}")
    print(f"\nTIME SCALE:")
    print(f"  Generations to equilibrium: ~{t_eq_mutation:,}")
    if t_eq_mutation < 10000:
        print(f"  → Relatively fast (thousands of generations)")
    else:
        print(f"  → Very slow (tens/hundreds of thousands of generations)")
    print(f"\nKEY INSIGHTS:")
    print(f"  → Mutation continuously creates deleterious alleles")
    print(f"  → Selection continuously removes them")
    print(f"  → Balance maintains allele at low but stable frequency")
    print(f"  → Can't eliminate disease alleles completely!")
    if selection_coeff > 0.1:
        print(f"  → Strong selection keeps frequency very low")
    else:
        print(f"  → Weak selection allows higher equilibrium frequency")
    print("="*70)
    
    fig.show()

# Interactive calculator
mu_slider = FloatLogSlider(value=1e-5, base=10, min=-7, max=-3, step=0.5,
                          description='Mutation (μ):')
s_slider = FloatSlider(value=0.01, min=0.001, max=0.5, step=0.001,
                      description='Selection (s):')
dom_dropdown = Dropdown(options=['recessive', 'dominant', 'additive'],
                       value='recessive', description='Dominance:')

display(HTML("<h3>🧬 Mutation-Selection Balance</h3>"))
interact(mutation_selection_balance, mutation_rate=mu_slider,
        selection_coeff=s_slider, dominance=dom_dropdown);
````

## Part 4: Evolutionary Forces Comparison

````python
def compare_forces(pop_size, sel_coeff, mig_rate, mut_rate, initial_p):
    """
    Compare relative strengths of all four evolutionary forces
    """
    
    q = 1 - initial_p
    p = initial_p
    
    # Calculate magnitude of each force (per generation)
    
    # Selection (assume directional against recessive)
    delta_q_selection = -sel_coeff * p * q**2
    
    # Drift (standard deviation, not deterministic change)
    sigma_drift = np.sqrt(p * q / (2 * pop_size))
    
    # Migration (assume q_migrant differs by 0.5)
    delta_q_migration = mig_rate * 0.5  # Assuming max difference
    
    # Mutation
    delta_q_mutation = mut_rate * p
    
    # Create comparison
    forces = {
        'Selection': abs(delta_q_selection),
        'Drift (σ)': sigma_drift,
        'Migration': delta_q_migration,
        'Mutation': delta_q_mutation
    }
    
    # Sort by magnitude
    sorted_forces = dict(sorted(forces.items(), key=lambda x: x[1], reverse=True))
    
    # Determine which dominates
    max_force = max(sorted_forces.values())
    dominant_force = [k for k, v in sorted_forces.items() if v == max_force][0]
    
    # Visualization
    fig = go.Figure()
    
    colors = {'Selection': '#E74C3C', 'Drift (σ)': '#F39C12',
             'Migration': '#3498DB', 'Mutation': '#2ECC71'}
    
    force_names = list(sorted_forces.keys())
    force_values = list(sorted_forces.values())
    force_colors = [colors[name] for name in force_names]
    
    fig.add_trace(go.Bar(
        x=force_names,
        y=force_values,
        marker_color=force_colors,
        text=[f"{v:.2e}" for v in force_values],
        textposition='outside'
    ))
    
    fig.update_layout(
        title='<b>Evolutionary Forces Comparison</b>',
        yaxis_title='Magnitude (|Δq| or σ per generation)',
        yaxis_type='log',
        height=500
    )
    
    # Print results
    print("\n" + "="*70)
    print("EVOLUTIONARY FORCES COMPARISON")
    print("="*70)
    print(f"\nPARAMETERS:")
    print(f"  Population size (N): {pop_size}")
    print(f"  Selection coefficient (s): {sel_coeff:.3f}")
    print(f"  Migration rate (m): {mig_rate:.3f}")
    print(f"  Mutation rate (μ): {mut_rate:.2e}")
    print(f"  Initial allele frequency (p): {initial_p:.2f}")
    print(f"\nFORCE MAGNITUDES (per generation):")
    print(f"  Selection: |Δq| = {abs(delta_q_selection):.4e}")
    print(f"  Drift: σ(Δq) = {sigma_drift:.4e}")
    print(f"  Migration: Δq = {delta_q_migration:.4e}")
    print(f"  Mutation: Δq = {delta_q_mutation:.4e}")
    print(f"\nRANKING (strongest → weakest):")
    for i, (force, magnitude) in enumerate(sorted_forces.items(), 1):
        print(f"  {i}. {force}: {magnitude:.4e}")
    print(f"\nDOMINANT FORCE: **{dominant_force}**")
    print(f"\nCRITICAL COMPARISONS:")
    print(f"  Selection vs Drift: {'Selection' if abs(delta_q_selection) > sigma_drift else 'Drift'} dominates")
    print(f"    (Ratio: {abs(delta_q_selection)/sigma_drift:.2f}:1)")
    print(f"  Selection vs Migration: {'Selection' if abs(delta_q_selection) > delta_q_migration else 'Migration'} dominates")
    print(f"    (Ratio: {abs(delta_q_selection)/delta_q_migration:.2f}:1)")
    print(f"  Drift vs Migration: {'Drift' if sigma_drift > delta_q_migration else 'Migration'} dominates")
    print(f"    (Ratio: {max(sigma_drift, delta_q_migration)/min(sigma_drift, delta_q_migration):.2f}:1)")
    print(f"\nIMPLICATIONS:")
    if dominant_force == 'Selection':
        print(f"  → Allele frequencies determined by fitness")
        print(f"  → Adaptation possible")
        print(f"  → Predictable evolution")
    elif dominant_force == 'Drift (σ)':
        print(f"  → Random changes in allele frequencies")
        print(f"  → No adaptation (unless lucky)")
        print(f"  → Unpredictable evolution")
    elif dominant_force == 'Migration':
        print(f"  → Gene flow homogenizes populations")
        print(f"  → Local adaptation prevented")
        print(f"  → Geographic uniformity")
    else:  # Mutation
        print(f"  → Very slow evolution")
        print(f"  → Long-term change only")
        print(f"  → (This is unusual - check parameters!)")
    print("="*70)
    
    fig.show()

# Interactive comparison
N_slider2 = IntSlider(value=100, min=10, max=10000, step=10,
                     description='Pop size (N):')
s_slider2 = FloatSlider(value=0.01, min=0.001, max=0.5, step=0.001,
                       description='Selection (s):')
m_slider2 = FloatSlider(value=0.01, min=0.001, max=0.5, step=0.001,
                       description='Migration (m):')
mu_slider2 = FloatLogSlider(value=1e-5, base=10, min=-7, max=-3, step=0.5,
                           description='Mutation (μ):')
p_slider3 = FloatSlider(value=0.5, min=0.1, max=0.9, step=0.1,
                       description='Initial p:')

display(HTML("<h3>⚖️ Forces Comparison Tool</h3>"))
interact(compare_forces, pop_size=N_slider2, sel_coeff=s_slider2,
        mig_rate=m_slider2, mut_rate=mu_slider2, initial_p=p_slider3);
````

## Part 5: Challenge Problems

### Challenge 1: Migration Between Two Populations 🌊

**Scenario**: Two equal-sized populations exchange migrants

- Population 1: q₁ = 0.2, N = 1000
- Population 2: q₂ = 0.8, N = 1000
- Migration rate: m = 0.10 (10% exchange)

**Questions**:
1. Calculate Δq for each population in generation 1
2. What is equilibrium frequency?
3. How many generations to reach 95% of equilibrium?

<details>
<summary>Solution</summary>

**Step 1: Calculate Δq**

Formula: Δq = m(q_m - q)

**Population 1** (receives from pop 2):
Δq₁ = 0.10(0.8 - 0.2) = 0.10(0.6) = **+0.06**
New q₁ = 0.2 + 0.06 = 0.26

**Population 2** (receives from pop 1):
Δq₂ = 0.10(0.2 - 0.8) = 0.10(-0.6) = **-0.06**
New q₂ = 0.8 - 0.06 = 0.74

**Symmetrical changes!** (equal-sized populations)

**Step 2: Equilibrium Frequency**

At equilibrium, both populations have the same frequency.

For equal-sized populations:
q_eq = (q₁ + q₂) / 2 = (0.2 + 0.8) / 2 = **0.5**

**Step 3: Time to 95% Equilibrium**

For each population approaching equilibrium:
q(t) = q_eq - (q_eq - q₀)(1 - m)^t

At 95% equilibrium:
0.95(q_eq - q₀) = (q_eq - q₀)(1 - (1-m)^t)
0.95 = 1 - (1-m)^t
(1-m)^t = 0.05

t = ln(0.05) / ln(1-m)
t = ln(0.05) / ln(0.90)
t = -2.996 / -0.105
t ≈ **29 generations**

**Verification**:
Starting difference: 0.6
After 29 gen: 0.6 × (0.90)^29 = 0.6 × 0.047 = 0.028
Remaining difference: 0.028 (5% of original) ✓

**Key Insights**:
- High migration (10%) homogenizes quickly
- Symmetrical when populations equal-sized
- Exponential approach to equilibrium
- Even small migration prevents divergence!
</details>

### Challenge 2: Cystic Fibrosis Equilibrium 🧬

**Given**: Cystic fibrosis (recessive, lethal)
- Observed disease frequency: 1 in 2,500 (0.0004)
- This implies q² = 0.0004, so q = 0.02

**Questions**:
1. What mutation rate maintains this frequency?
2. Is this a realistic mutation rate?
3. Why is CF so common if it's lethal?

<details>
<summary>Solution</summary>

**Step 1: Calculate Required Mutation Rate**

For recessive lethal at mutation-selection balance:
q̂ = √(μ/s)

With s = 1 (lethal):
q̂ = √μ

Observed: q = 0.02

Therefore:
0.02 = √μ
μ = (0.02)²
μ = **0.0004 = 4 × 10⁻⁴**

**Step 2: Is This Realistic?**

**NO!** This is way too high!

Typical mutation rates:
- Per base pair: μ ≈ 10⁻⁹
- Per gene: μ ≈ 10⁻⁶ to 10⁻⁵

Required: μ = 4 × 10⁻⁴ is **40× to 400× higher than typical!**

**Step 3: Why So Common?**

**Heterozygote Advantage Hypothesis!**

Historically, CF carriers (Aa) may have had resistance to:
- Cholera (toxin blocks chloride channels, but CF defect prevents this)
- Typhoid fever
- Tuberculosis

**Alternative explanation**:

With heterozygote advantage (s₁, s₂):
q̂ = s₁ / (s₁ + s₂)

If:
- s₁ (against AA) = 0.02 (2% from disease)
- s₂ (against aa) = 1.00 (100%, lethal)

Then:
q̂ = 0.02 / (0.02 + 1.00) = 0.02 / 1.02 ≈ **0.0196 ≈ 0.02** ✓

This matches observed frequency!

**Evidence**:
- CF common in European descent (where cholera was endemic)
- CF rare in populations without historical cholera
- CFTR defect does provide cholera resistance in vitro

**Modern situation**:
- No cholera in developed world
- No longer heterozygote advantage
- CF frequency should slowly decline
- But takes thousands of generations!

**Key Lesson**:
When disease frequencies don't match mutation-selection balance,
consider heterozygote advantage! (Like sickle cell)
</details>

### Challenge 3: Which Force Dominates? ⚖️

**Scenario**: Small island population
- N = 50
- Slightly deleterious allele: s = 0.02
- Migration from mainland: m = 0.05
- Current frequency: q = 0.1
- Mainland frequency: q_m = 0.3

**Question**: Which force dominates: selection, drift, or migration?

<details>
<summary>Solution</summary>

**Calculate magnitude of each force:**

**Selection** (against recessive, p=0.9, q=0.1):
Δq_selection = -spq²
= -0.02 × 0.9 × 0.01
= -0.00018
|Δq_selection| = **0.00018**

**Drift** (standard deviation):
σ(Δq) = √(pq/2N)
= √(0.9 × 0.1 / 100)
= √0.0009
= **0.03**

**Migration**:
Δq_migration = m(q_m - q)
= 0.05(0.3 - 0.1)
= 0.05(0.2)
= **0.01**

**Comparison**:
- Drift (σ): 0.03
- Migration: 0.01
- Selection: 0.00018

**Ranking**: Drift > Migration >> Selection

**Ratios**:
- Drift / Selection = 0.03 / 0.00018 = **167×** stronger!
- Migration / Selection = 0.01 / 0.00018 = **56×** stronger!
- Drift / Migration = 0.03 / 0.01 = **3×** stronger

**Answer: DRIFT dominates, with Migration second**

**Implications**:

**Selection irrelevant** (N too small):
- Even though allele is deleterious (s=0.02)
- Selection too weak to overcome drift
- Criterion: s > 1/2N = 1/100 = 0.01
- Since s = 0.02 > 0.01, selection SHOULD work
- But drift variance (0.03) still larger!

**Migration matters**:
- Introduces allele from mainland
- Δq = +0.01 per generation
- Opposes any selection or drift toward lower q

**Drift dominates both**:
- Random changes (±0.03) overwhelm
- Unpredictable trajectory
- Could go up or down

**Prediction**:
Allele frequency will:
- Fluctuate randomly (drift)
- With slight upward trend (migration)
- Ignoring fitness effects (selection too weak)
- Eventually may drift to fixation or loss

**Management implications** (if we wanted to remove this allele):
1. **Increase N**: If N > 500, selection would dominate
2. **Stop migration**: Eliminates source of allele
3. **Accept reality**: Can't control drift in small populations

**Key Lesson**:
In small populations, drift dominates everything except very strong selection or very high migration!
</details>

### Migration Properties

**1. Homogenizing force**
- Makes populations similar
- Opposes local adaptation
- Even small migration effective (Nm > 1)

**2. Speed depends on m**
- High m (>10%): Rapid homogenization
- Low m (<1%): Slow
- Time to 95% equilibrium: t ≈ -ln(0.05)/ln(1-m)

**3. Can maintain maladaptive alleles**
- If m > s, gene flow wins
- Prevents local adaptation

### Mutation Properties

**1. Very slow**
- Typical μ = 10⁻⁶ to 10⁻⁵
- Takes ~1/μ generations for significant change
- 100,000+ generations!

**2. Creates variation**
- Only source of new alleles
- Essential for long-term evolution
- Opposes selection and drift

**3. Maintains deleterious alleles**
- Mutation-selection balance
- Can't eliminate genetic disease
- Genetic load unavoidable
