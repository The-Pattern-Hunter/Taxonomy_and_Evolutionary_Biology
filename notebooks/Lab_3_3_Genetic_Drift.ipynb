## Part 1: Genetic Drift Theory

### What is Genetic Drift?

**Definition:**
Random changes in allele frequencies due to sampling error in finite populations.

**Key Insight:**
Even without selection, mutation, or migration, allele frequencies change randomly from generation to generation!

### The Coin Flip Analogy

**Fair coin:** p(heads) = 0.5, p(tails) = 0.5

**Flip 10 times:** Might get 6 heads, 4 tails (60% vs 40%)  
**Flip 1000 times:** Very close to 500 heads, 500 tails (50% vs 50%)

**Small samples → large random deviations**  
**Large samples → small random deviations**

Same principle applies to reproduction!

### Mechanism of Drift

**Population reproduction is random sampling:**

1. **Parent generation:** Has certain allele frequencies (e.g., p = 0.5)
2. **Random mating:** Produces gametes (like drawing from a hat)
3. **Offspring generation:** Sample of gametes becomes next generation
4. **Result:** Offspring allele frequencies differ from parents by chance!

**Example:**
```
Population: 10 diploid individuals (20 alleles total)
Current: 10 A alleles, 10 a alleles (p = q = 0.5)

Next generation (random sample):
Trial 1: 12 A, 8 a (p = 0.60) ← Drift!
Trial 2: 9 A, 11 a (p = 0.45) ← Drift!
Trial 3: 10 A, 10 a (p = 0.50) ← No change (by chance)
```

### Mathematical Properties

**Variance in allele frequency change:**
## σ²(Δq) = pq / 2N

Where:
- σ²(Δq) = variance in change of q
- p, q = allele frequencies
- N = population size (diploid individuals)
- 2N = total number of gene copies

**Key insights:**
- Variance is **inversely proportional to N**
- Small populations → large variance → strong drift
- Large populations → small variance → weak drift
- Maximum variance when p = q = 0.5

**Standard deviation:**
## σ(Δq) = √(pq / 2N)

**Example:**
```
N = 10, p = q = 0.5:
σ(Δq) = √(0.25 / 20) = √0.0125 = 0.112 (11.2% per generation!)

N = 1000, p = q = 0.5:
σ(Δq) = √(0.25 / 2000) = √0.000125 = 0.011 (1.1% per generation)
```

### Loss of Heterozygosity

**Drift reduces genetic diversity over time!**

**Heterozygosity decay:**
## H_t = H_0 (1 - 1/2N)^t

Where:
- H_t = heterozygosity at generation t
- H_0 = initial heterozygosity (usually 2pq)
- N = population size
- t = generations

**Rate of heterozygosity loss:**
## ΔH/H = -1/2N per generation

**Example:**
```
N = 50, H_0 = 0.5
After 100 generations:
H_100 = 0.5 × (1 - 1/100)^100 = 0.5 × 0.366 = 0.183
Lost 63% of heterozygosity!
```

### Probability of Fixation

**Eventually, one allele will be fixed (frequency = 1.0) by drift!**

**For neutral alleles:**
## P(fixation) = p

Where p = current frequency of the allele

**Key insight:**
- Probability of fixation equals current frequency!
- Common alleles more likely to be fixed
- Rare alleles likely to be lost
- But rare alleles CAN be fixed by drift (just unlikely)

**Example:**
```
Allele at p = 0.60:
P(A fixed) = 0.60 (60% chance)
P(a fixed) = 0.40 (40% chance)
```

**Time to fixation (if it happens):**
## t ≈ 4N generations

**Longer for larger populations!**

### Effective Population Size (N_e)

**Not all individuals contribute equally to next generation!**

**N_e = Effective population size**
- The size of an "ideal" population experiencing the same amount of drift
- Usually N_e < N (census size)

**Factors reducing N_e:**
1. **Unequal sex ratio:** N_e = 4N_m N_f / (N_m + N_f)
2. **Variation in reproductive success:** Some breed more than others
3. **Non-random mating:** Inbreeding, mate choice
4. **Overlapping generations:** Age structure effects

**Example (unequal sex ratio):**
```
100 males, 10 females:
N_e = 4(100)(10) / (100+10) = 4000/110 = 36.4
Much smaller than N = 110!
```

### Founder Effect

**Definition:**
Loss of genetic variation when a new population is established by a small number of individuals.

**Mechanism:**
- Large source population with frequencies p, q
- Small group colonizes new area (e.g., 5 individuals)
- By chance, founder allele frequencies differ from source
- New population starts with reduced diversity

**Classic examples:**

**1. Amish populations:**
- Founded by ~200 individuals
- High frequency of Ellis-van Creveld syndrome
- Due to founder carrying rare allele

**2. Tristan da Cunha:**
- Island with ~300 people
- Founded by 15 individuals (1817)
- High frequency of retinitis pigmentosa

**3. Cheetahs:**
- Extremely low genetic diversity
- Severe bottleneck ~10,000 years ago
- Can transplant skin between unrelated individuals!

### Bottleneck Effect

**Definition:**
Sharp reduction in population size followed by recovery.

**Phases:**
1. **Before:** Large population, high diversity
2. **Bottleneck:** Population crashes (disease, disaster, etc.)
3. **After:** Population recovers in size BUT diversity lost

**Genetic consequences:**
- Loss of rare alleles (likely lost in bottleneck)
- Reduced heterozygosity
- Random changes in allele frequencies
- Effects permanent even after population recovers!

**Examples:**

**1. Northern elephant seals:**
- Hunted to ~20 individuals in 1890s
- Now >100,000 individuals
- BUT: Almost zero genetic diversity
- All descended from ~20 survivors

**2. Cheetahs (as above)**

**3. Greater prairie chicken:**
- Millions → <50 in Illinois
- Loss of fertility from inbreeding
- Genetic rescue by translocation

### Drift vs Selection

**When does drift matter?**

**Drift dominates when:**
## |s| < 1/2N

If selection coefficient |s| < 1/(2N), drift is stronger than selection!

**Examples:**
```
N = 100:
If |s| < 0.005 (0.5%), drift dominates

N = 10,000:
If |s| < 0.00005 (0.005%), drift dominates
```

**Implications:**
- **Small populations:** Drift overpowers even moderate selection
- **Large populations:** Selection overpowers drift for all but tiniest s
- **Nearly neutral alleles:** Always drift-dominated

**Key insight:**
In small populations, slightly deleterious alleles can drift to fixation!
This is BAD for conservation (accumulation of bad alleles).

### Wright-Fisher Model

**Standard model of genetic drift:**

**Assumptions:**
1. Constant population size (N)
2. Random mating
3. Non-overlapping generations
4. Neutral alleles (no selection)

**Process:**
Each generation, draw 2N alleles randomly (with replacement) from previous generation's gene pool.

**Mathematical treatment:**
Binomial sampling: If current frequency is p, next generation has:
## i = Binomial(2N, p) copies of A allele
## p' = i / 2N

### Why Drift Matters

**1. Conservation biology:**
- Small populations lose diversity → extinction risk
- Minimum viable population (MVP) concepts
- Genetic rescue through translocation

**2. Molecular evolution:**
- Most mutations neutral or nearly neutral
- Drift drives fixation of neutral mutations
- Molecular clock based on neutral drift

**3. Human populations:**
- Founder effects in isolated populations
- Disease alleles at high frequency
- Loss of diversity in bottlenecks

**4. Experimental evolution:**
- Small lab populations experience strong drift
- Can obscure selection signals
- Need large populations to study selection

### The Neutral Theory

**Motoo Kimura (1968):**
Most molecular evolution is due to drift, not selection!

**Key claims:**
- Most mutations are neutral (no fitness effect)
- Neutral mutations fixed by drift at rate = mutation rate
- Explains molecular clock (constant rate of evolution)

**Implications:**
- Drift, not selection, drives most DNA sequence changes
- Selection important for adaptive evolution
- But quantitatively, drift fixes more mutations

**Modern view:**
Both drift and selection important, depends on:
- Type of mutation (synonymous vs nonsynonymous)
- Population size
- Strength of selection

## Part 3: Genetic Drift Simulator

````python
def drift_simulator(pop_size, initial_p, num_generations, num_replicates):
    """
    Simulate genetic drift using Wright-Fisher model
    """
    
    # Storage for all replicates
    all_trajectories = []
    fixed_A = 0  # Count fixations
    fixed_a = 0
    still_varying = 0
    
    # Run multiple replicates
    for rep in range(num_replicates):
        p = initial_p
        trajectory = [p]
        
        for gen in range(num_generations):
            # Binomial sampling (Wright-Fisher model)
            # Sample 2N alleles with probability p
            num_A_alleles = np.random.binomial(2 * pop_size, p)
            p = num_A_alleles / (2 * pop_size)
            trajectory.append(p)
            
            # Stop if fixed
            if p == 0 or p == 1:
                break
        
        # Pad trajectory to full length if stopped early
        final_p = trajectory[-1]
        while len(trajectory) < num_generations + 1:
            trajectory.append(final_p)
        
        all_trajectories.append(trajectory)
        
        # Count outcomes
        if final_p == 1:
            fixed_A += 1
        elif final_p == 0:
            fixed_a += 1
        else:
            still_varying += 1
    
    # Calculate statistics
    all_trajectories = np.array(all_trajectories)
    mean_trajectory = np.mean(all_trajectories, axis=0)
    std_trajectory = np.std(all_trajectories, axis=0)
    
    # Theoretical variance
    q = 1 - initial_p
    theoretical_var = (initial_p * q) / (2 * pop_size)
    theoretical_std = np.sqrt(theoretical_var)
    
    # Heterozygosity loss
    initial_H = 2 * initial_p * (1 - initial_p)
    gens = np.arange(num_generations + 1)
    expected_H = initial_H * (1 - 1/(2*pop_size))**gens
    
    # Observed heterozygosity
    observed_H = []
    for gen in range(num_generations + 1):
        p_values = all_trajectories[:, gen]
        H_values = 2 * p_values * (1 - p_values)
        observed_H.append(np.mean(H_values))
    
    # Visualization
    fig = make_subplots(
        rows=2, cols=2,
        subplot_titles=(f'Individual Replicates (n={min(20, num_replicates)} shown)',
                       'Mean Allele Frequency ± 1 SD',
                       'Loss of Heterozygosity',
                       'Final Outcomes'),
        specs=[[{'type': 'scatter'}, {'type': 'scatter'}],
               [{'type': 'scatter'}, {'type': 'bar'}]]
    )
    
    # 1. Individual trajectories (show up to 20)
    for i in range(min(20, num_replicates)):
        fig.add_trace(go.Scatter(
            x=gens, y=all_trajectories[i],
            mode='lines',
            line=dict(width=1, color='rgba(100,100,100,0.3)'),
            showlegend=False
        ), row=1, col=1)
    
    # 2. Mean ± SD
    fig.add_trace(go.Scatter(
        x=gens, y=mean_trajectory,
        mode='lines', name='Mean',
        line=dict(color='#E74C3C', width=3)
    ), row=1, col=2)
    fig.add_trace(go.Scatter(
        x=gens, y=mean_trajectory + std_trajectory,
        mode='lines', name='+1 SD',
        line=dict(color='#E74C3C', width=1, dash='dash'),
        showlegend=False
    ), row=1, col=2)
    fig.add_trace(go.Scatter(
        x=gens, y=mean_trajectory - std_trajectory,
        mode='lines', name='-1 SD',
        line=dict(color='#E74C3C', width=1, dash='dash'),
        showlegend=False
    ), row=1, col=2)
    
    # 3. Heterozygosity
    fig.add_trace(go.Scatter(
        x=gens, y=expected_H,
        mode='lines', name='Expected H',
        line=dict(color='#3498DB', width=2, dash='dash')
    ), row=2, col=1)
    fig.add_trace(go.Scatter(
        x=gens, y=observed_H,
        mode='lines', name='Observed H',
        line=dict(color='#2ECC71', width=3)
    ), row=2, col=1)
    
    # 4. Outcomes
    outcomes = ['A fixed<br>(p=1)', 'Still varying', 'a fixed<br>(p=0)']
    counts = [fixed_A, still_varying, fixed_a]
    colors = ['#3498DB', '#F39C12', '#E74C3C']
    
    fig.add_trace(go.Bar(
        x=outcomes, y=counts,
        marker_color=colors,
        text=counts,
        textposition='outside',
        showlegend=False
    ), row=2, col=2)
    
    fig.update_xaxes(title_text="Generation", row=1, col=1)
    fig.update_xaxes(title_text="Generation", row=1, col=2)
    fig.update_xaxes(title_text="Generation", row=2, col=1)
    
    fig.update_yaxes(title_text="Frequency (p)", range=[0, 1], row=1, col=1)
    fig.update_yaxes(title_text="Frequency (p)", range=[0, 1], row=1, col=2)
    fig.update_yaxes(title_text="Heterozygosity", row=2, col=1)
    fig.update_yaxes(title_text="Count", row=2, col=2)
    
    fig.update_layout(height=700, title_text='<b>Genetic Drift Simulation</b>')
    
    # Print results
    print("\n" + "="*70)
    print("GENETIC DRIFT SIMULATION RESULTS")
    print("="*70)
    print(f"\nPARAMETERS:")
    print(f"  Population size (N): {pop_size}")
    print(f"  Initial p: {initial_p:.3f}")
    print(f"  Generations: {num_generations}")
    print(f"  Replicates: {num_replicates}")
    print(f"\nTHEORETICAL PREDICTIONS:")
    print(f"  Variance σ²(Δq) = pq/2N = {theoretical_var:.6f}")
    print(f"  Std deviation σ(Δq) = {theoretical_std:.4f}")
    print(f"  Expected change per generation: ±{theoretical_std:.4f}")
    print(f"\nOBSERVED RESULTS:")
    print(f"  Mean final p: {mean_trajectory[-1]:.3f}")
    print(f"  SD final p: {std_trajectory[-1]:.3f}")
    print(f"\nHETEROZYGOSITY:")
    print(f"  Initial: H_0 = {initial_H:.4f}")
    print(f"  Expected final: H_{num_generations} = {expected_H[-1]:.4f}")
    print(f"  Observed final: {observed_H[-1]:.4f}")
    print(f"  Loss: {(1 - observed_H[-1]/initial_H)*100:.1f}%")
    print(f"\nFIXATION:")
    print(f"  A fixed (p=1): {fixed_A}/{num_replicates} ({fixed_A/num_replicates*100:.1f}%)")
    print(f"  a fixed (p=0): {fixed_a}/{num_replicates} ({fixed_a/num_replicates*100:.1f}%)")
    print(f"  Still varying: {still_varying}/{num_replicates} ({still_varying/num_replicates*100:.1f}%)")
    print(f"\nEXPECTED FIXATION PROBABILITY: {initial_p:.3f} (= initial p)")
    if fixed_A + fixed_a > 0:
        observed_prob = fixed_A / (fixed_A + fixed_a)
        print(f"OBSERVED FIXATION PROBABILITY: {observed_prob:.3f}")
    print(f"\nKEY INSIGHTS:")
    if pop_size < 50:
        print(f"  → STRONG drift (small population)")
        print(f"  → Rapid loss of diversity")
        print(f"  → Fixation likely within {4*pop_size} generations")
    elif pop_size < 500:
        print(f"  → MODERATE drift")
        print(f"  → Gradual loss of diversity")
    else:
        print(f"  → WEAK drift (large population)")
        print(f"  → Slow loss of diversity")
        print(f"  → Would take ~{4*pop_size} generations for fixation")
    print("="*70)
    
    fig.show()

# Interactive simulator
N_slider = IntSlider(value=50, min=10, max=500, step=10,
                    description='Pop size (N):')
p_slider = FloatSlider(value=0.5, min=0.1, max=0.9, step=0.1,
                      description='Initial p:')
gen_slider = IntSlider(value=100, min=50, max=500, step=50,
                      description='Generations:')
rep_slider = IntSlider(value=50, min=10, max=100, step=10,
                      description='Replicates:')

display(HTML("<h3>🎲 Genetic Drift Simulator</h3>"))
display(HTML("<p>Simulate random changes in allele frequency (Wright-Fisher model):</p>"))
interact(drift_simulator, pop_size=N_slider, initial_p=p_slider,
        num_generations=gen_slider, num_replicates=rep_slider);
````

## Part 4: Bottleneck & Founder Effect Simulator

````python
def bottleneck_simulator(initial_N, bottleneck_N, recovery_N, bottleneck_duration,
                         recovery_time, initial_p):
    """
    Simulate population bottleneck and recovery
    """
    
    # Phase 1: Pre-bottleneck
    phase1_gens = 50
    N1 = initial_N
    p = initial_p
    trajectory_p = [p]
    trajectory_N = [N1]
    
    for gen in range(phase1_gens):
        num_A = np.random.binomial(2 * N1, p)
        p = num_A / (2 * N1)
        trajectory_p.append(p)
        trajectory_N.append(N1)
    
    # Phase 2: Bottleneck
    N2 = bottleneck_N
    for gen in range(bottleneck_duration):
        num_A = np.random.binomial(2 * N2, p)
        p = num_A / (2 * N2)
        trajectory_p.append(p)
        trajectory_N.append(N2)
    
    # Phase 3: Recovery
    # Gradual increase in N
    for gen in range(recovery_time):
        progress = gen / recovery_time
        N3 = int(N2 + (recovery_N - N2) * progress)
        num_A = np.random.binomial(2 * N3, p)
        p = num_A / (2 * N3)
        trajectory_p.append(p)
        trajectory_N.append(N3)
    
    # Phase 4: Post-recovery
    phase4_gens = 50
    N4 = recovery_N
    for gen in range(phase4_gens):
        num_A = np.random.binomial(2 * N4, p)
        p = num_A / (2 * N4)
        trajectory_p.append(p)
        trajectory_N.append(N4)
    
    # Calculate heterozygosity
    trajectory_H = [2 * p_val * (1 - p_val) for p_val in trajectory_p]
    
    # Statistics
    initial_H = 2 * initial_p * (1 - initial_p)
    final_H = trajectory_H[-1]
    H_loss_pct = (1 - final_H / initial_H) * 100 if initial_H > 0 else 0
    
    # Visualization
    fig = make_subplots(
        rows=2, cols=1,
        subplot_titles=('Population Size Over Time', 'Allele Frequency & Heterozygosity'),
        specs=[[{'type': 'scatter'}], [{'type': 'scatter'}]],
        vertical_spacing=0.15
    )
    
    gens = list(range(len(trajectory_N)))
    
    # Mark phases
    phase1_end = phase1_gens
    phase2_end = phase1_end + bottleneck_duration
    phase3_end = phase2_end + recovery_time
    
    # 1. Population size
    fig.add_trace(go.Scatter(
        x=gens, y=trajectory_N,
        mode='lines',
        line=dict(color='#3498DB', width=3),
        fill='tozeroy',
        fillcolor='rgba(52, 152, 219, 0.2)',
        name='Population Size'
    ), row=1, col=1)
    
    # Mark bottleneck
    fig.add_vrect(x0=phase1_end, x1=phase2_end,
                  fillcolor='rgba(231, 76, 60, 0.2)',
                  layer='below', line_width=0,
                  annotation_text="Bottleneck",
                  annotation_position="top",
                  row=1, col=1)
    
    # 2. Allele frequency
    fig.add_trace(go.Scatter(
        x=gens, y=trajectory_p,
        mode='lines',
        line=dict(color='#E74C3C', width=2),
        name='Allele freq (p)'
    ), row=2, col=1)
    
    # 3. Heterozygosity
    fig.add_trace(go.Scatter(
        x=gens, y=trajectory_H,
        mode='lines',
        line=dict(color='#2ECC71', width=2),
        name='Heterozygosity (H)'
    ), row=2, col=1)
    
    fig.add_vrect(x0=phase1_end, x1=phase2_end,
                  fillcolor='rgba(231, 76, 60, 0.2)',
                  layer='below', line_width=0,
                  row=2, col=1)
    
    fig.update_xaxes(title_text="Generation", row=2, col=1)
    fig.update_yaxes(title_text="N (individuals)", row=1, col=1)
    fig.update_yaxes(title_text="Frequency / H", range=[0, 1], row=2, col=1)
    
    fig.update_layout(height=700,
                     title_text='<b>Population Bottleneck Simulation</b>')
    
    # Print results
    print("\n" + "="*70)
    print("BOTTLENECK SIMULATION RESULTS")
    print("="*70)
    print(f"\nPHASE 1 - PRE-BOTTLENECK:")
    print(f"  Population size: {initial_N}")
    print(f"  Duration: {phase1_gens} generations")
    print(f"  Initial p: {initial_p:.3f}")
    print(f"  Initial H: {initial_H:.4f}")
    print(f"\nPHASE 2 - BOTTLENECK:")
    print(f"  Population crashed to: {bottleneck_N} ({'%.1f'%(bottleneck_N/initial_N*100)}% of original)")
    print(f"  Duration: {bottleneck_duration} generations")
    print(f"  Drift strength: σ(Δq) = {np.sqrt(initial_p*(1-initial_p)/(2*bottleneck_N)):.4f}")
    print(f"\nPHASE 3 - RECOVERY:")
    print(f"  Recovery to: {recovery_N}")
    print(f"  Duration: {recovery_time} generations")
    print(f"\nPHASE 4 - POST-RECOVERY:")
    print(f"  Stable population: {recovery_N}")
    print(f"  Duration: {phase4_gens} generations")
    print(f"\nGENETIC CONSEQUENCES:")
    print(f"  Initial p: {initial_p:.3f} → Final p: {trajectory_p[-1]:.3f}")
    print(f"  Change in p: {(trajectory_p[-1] - initial_p):.3f}")
    print(f"  Initial H: {initial_H:.4f} → Final H: {final_H:.4f}")
    print(f"  Heterozygosity LOST: {H_loss_pct:.1f}%")
    print(f"\nKEY INSIGHTS:")
    if H_loss_pct > 50:
        print(f"  → SEVERE diversity loss!")
        print(f"  → Population at risk (inbreeding depression)")
        print(f"  → Genetic rescue recommended")
    elif H_loss_pct > 20:
        print(f"  → MODERATE diversity loss")
        print(f"  → Monitoring recommended")
    else:
        print(f"  → MINOR diversity loss")
    if abs(trajectory_p[-1] - initial_p) > 0.2:
        print(f"  → Large change in allele frequency (drift effect)")
    print(f"\nIMPORTANT: Diversity loss is PERMANENT!")
    print(f"  Population size recovered but genetic diversity did NOT.")
    print("="*70)
    
    fig.show()

# Interactive simulator
initial_N_slider = IntSlider(value=1000, min=100, max=5000, step=100,
                            description='Initial N:')
bottleneck_N_slider = IntSlider(value=20, min=5, max=100, step=5,
                               description='Bottleneck N:')
recovery_N_slider = IntSlider(value=1000, min=100, max=5000, step=100,
                             description='Recovery N:')
duration_slider = IntSlider(value=10, min=5, max=50, step=5,
                           description='Bottleneck gens:')
recovery_slider = IntSlider(value=50, min=10, max=100, step=10,
                           description='Recovery gens:')
p_slider2 = FloatSlider(value=0.5, min=0.1, max=0.9, step=0.1,
                       description='Initial p:')

display(HTML("<h3>💥 Bottleneck Simulator</h3>"))
display(HTML("<p>Model population crash and recovery:</p>"))
interact(bottleneck_simulator, initial_N=initial_N_slider,
        bottleneck_N=bottleneck_N_slider, recovery_N=recovery_N_slider,
        bottleneck_duration=duration_slider, recovery_time=recovery_slider,
        initial_p=p_slider2);
````

**Question:** Three populations (N=10, N=100, N=1000) all start with p = 0.5

1. Calculate expected σ(Δq) for each
2. After 100 generations, what % heterozygosity is lost?
3. Which populations most likely to lose the 'a' allele by chance?

<details>
<summary>Solution</summary>

**Step 1: Standard Deviation of Δq**

Formula: σ(Δq) = √(pq/2N)

With p = q = 0.5:

**N = 10:**
σ(Δq) = √(0.25 / 20) = √0.0125 = **0.112**
→ 11.2% change per generation!

**N = 100:**
σ(Δq) = √(0.25 / 200) = √0.00125 = **0.035**
→ 3.5% change per generation

**N = 1000:**
σ(Δq) = √(0.25 / 2000) = √0.000125 = **0.011**
→ 1.1% change per generation

**Ratio:** N=10 is 10× stronger drift than N=100, 100× stronger than N=1000!

**Step 2: Heterozygosity Loss**

Formula: H_t = H_0(1 - 1/2N)^t

Initial: H_0 = 2pq = 2(0.5)(0.5) = 0.5

After t = 100 generations:

**N = 10:**
H_100 = 0.5 × (1 - 1/20)^100 = 0.5 × (0.95)^100
H_100 = 0.5 × 0.00592 = **0.003**
Loss: (0.5 - 0.003)/0.5 = **99.4% LOST!**

**N = 100:**
H_100 = 0.5 × (1 - 1/200)^100 = 0.5 × (0.995)^100
H_100 = 0.5 × 0.606 = **0.303**
Loss: (0.5 - 0.303)/0.5 = **39.4% lost**

**N = 1000:**
H_100 = 0.5 × (1 - 1/2000)^100 = 0.5 × (0.9995)^100
H_100 = 0.5 × 0.951 = **0.476**
Loss: (0.5 - 0.476)/0.5 = **4.9% lost**

**Summary:**
- N=10: Almost all diversity lost
- N=100: About 1/3 diversity lost
- N=1000: Minimal diversity lost

**Step 3: Probability of Allele Loss**

Time to fixation: t ≈ 4N generations

**N = 10:** t ≈ 40 generations
→ After 100 gen (2.5× time to fixation), very likely ONE allele fixed
→ HIGH probability of losing either A or a

**N = 100:** t ≈ 400 generations
→ After 100 gen (1/4 time to fixation), fixation unlikely
→ MODERATE probability, but most populations still varying

**N = 1000:** t ≈ 4000 generations
→ After 100 gen (1/40 time to fixation), fixation very unlikely
→ LOW probability of losing allele

**Key Insight:**
Small populations lose diversity RAPIDLY through drift!
This is why conservation biology focuses on maintaining large N_e.

### Key Concepts

✅ **Genetic Drift** - Random changes in allele frequency  
✅ **Sampling Error** - Mechanism of drift in finite populations  
✅ **Variance**: σ²(Δq) = pq/2N (inversely proportional to N)  
✅ **Heterozygosity Loss**: H_t = H_0(1 - 1/2N)^t  
✅ **Fixation Probability** = current allele frequency  
✅ **Time to Fixation** ≈ 4N generations  

### Fundamental Equations

**Variance in Δq:**
## σ²(Δq) = pq / 2N

**Heterozygosity decay:**
## H_t = H_0 (1 - 1/2N)^t

**Effective population size (unequal sex ratio):**
## N_e = 4N_m N_f / (N_m + N_f)

**Drift vs Selection:**
## Drift dominates when |s| < 1/2N

### Properties of Drift

**1. Random (not directed)**
- No adaptation produced
- Direction unpredictable
- Can fix deleterious alleles

**2. Stronger in small populations**
- N=10: σ(Δq) = 11%/gen
- N=100: σ(Δq) = 3.5%/gen
- N=1000: σ(Δq) = 1.1%/gen

**3. Reduces genetic diversity**
- Heterozygosity declines
- Alleles lost by chance
- Eventually → fixation

**4. Permanent effects**
- Once allele lost, gone forever
- Population recovery ≠ diversity recovery
- Mutation too slow to restore

### Classic Examples

**Northern Elephant Seal:**
- N crashed to ~20 in 1890s
- Now 100,000 but almost zero diversity
- Can accept skin grafts between unrelated individuals

**Cheetahs:**
- Bottleneck ~10,000 years ago
- Extremely low diversity
- All individuals essentially clones

**Amish (Ellis-van Creveld):**
- Founded by ~200 individuals
- Rare allele became common (founder effect)
- 1 in 200 affected vs 1 in 60,000 normally

### Drift vs Selection

**Selection dominates when:**
- Large populations (N > 500)
- Strong selection (|s| > 0.01)
- s × N > 1

**Drift dominates when:**
- Small populations (N < 50)
- Weak selection (|s| < 1/2N)
- Nearly neutral alleles

**Consequences:**
- Small pops: Slightly bad alleles can fix
- Large pops: Even weak selection effective
- Conservation: Need N_e > 50-100 minimum

### Conservation Implications

**Minimum Viable Population (MVP):**
- N_e > 50: Avoid inbreeding depression (short-term)
- N_e > 500: Maintain adaptive potential (long-term)
- N_e > 5000: Preserve evolutionary potential

**Risks of small populations:**
1. Loss of genetic diversity
2. Inbreeding depression
3. Accumulation of deleterious alleles
4. Reduced disease resistance
5. Loss of adaptive potential

**Management strategies:**
- Maintain large N_e
- Genetic rescue (translocation)
- Prevent bottlenecks
- Monitor diversity
- Manage breeding programs

### The Neutral Theory

**Kimura (1968):**
Most molecular evolution driven by drift, not selection!

**Key ideas:**
- Most mutations neutral (no fitness effect)
- Fixed by drift at rate = mutation rate
- Explains molecular clock

**Modern synthesis:**
- Drift drives most DNA changes
- Selection drives adaptation
- Both important, context-dependent

### Drift in Different Contexts

**Molecular evolution:**
- Neutral mutations fixed by drift
- Molecular clock
- Phylogenetic inference

**Conservation:**
- Loss of diversity in small populations
- Inbreeding depression
- Minimum viable population

**Human genetics:**
- Founder effects (Amish, Finnish, etc.)
- Population bottlenecks
- Disease allele frequencies

**Experimental evolution:**
- Small lab populations
- Confounds selection studies
- Need large N to study selection

### The Big Picture

**Drift is evolution by CHANCE, selection is evolution by CHOICE**

**Key differences:**

| Property | Drift | Selection |
|----------|-------|----------|
| Direction | Random | Toward higher fitness |
| Strength | ∝ 1/N | ∝ s |
| Adaptation | No | Yes |
| Diversity | Reduces | Can maintain (balancing) |
| Predictable | No | Yes |

**Together:**
- Drift and selection operate simultaneously
- Relative strength depends on N and s
- Small pops: Drift wins
- Large pops: Selection wins

**Conservation lesson:**
**Keep populations LARGE to let selection work effectively!**
